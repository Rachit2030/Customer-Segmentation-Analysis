###  What is Customer Segmentation?

> 1 ) Customer segmentation is the process of dividing customers into groups based on common characterstics so that the company can market each group effectively and appropriately.<br>
> 2 ) The goal of segmenting customers is to decide how to relate customers in each segment in order to maximize the value of each customer to the business.<br>
> 3 ) A customer segmentation analysis allows marketers to identify descrete groups of customers with high degree of accuracy based on demographic, behavioural and other indicators.<br>

### Model Fitting

* We selected the features that we wanted to keep for clustering and find the optimal number of clusters using elbow method<br>

```python
elements = [
    "Education",
    "Marital_Status",
    "Income",
    "Wine",
    "Age",
    "Customer_Seniority",
    "Children",
    "Total",
    "Total_Campaign"
]

elbow = KElbowVisualizer(KMeans(),k = 8)
elbow.fit(df[elements])
elbow.show()
plt.show()
```

```python
kmeancluster = KMeans(n_clusters=4)
df["kcluster"] = kmeancluster.fit_predict(df[elements])
```

```python
sns.countplot(data=df, x="kcluster", palette="tab20")
plt.title("Cluster Distribution")
plt.xlabel("Cluster")
plt.show()
```

### Data Exploration

* **Plot Our Clusters between Income and Total money Spent in order to visualise the relationship** 

```python
plt.figure(figsize=(10,8))
pl = sns.scatterplot(data = df,x=df["Total"],y=df["Income"],hue = df["kcluster"],palette = "tab20")
pl.set_title("Clusters w.r.t Income and Spending")
plt.legend()
plt.show()
```
